# Find null values

```python
df.isnull()
```

# Drop umpire3 column as it contains more than 75% of null values

```python
df.drop("umpire3",axis=1,inplace=True)
df.head()
```

# Drop null values of city , winner ,player_of_match,umpire1,umpire2 column

```python
df.dropna(subset=["city", "winner", "player_of_match", "umpire1", "umpire2"], inplace=True)
df.head()
```

# which city hosted most number of matches?
draw bar plot and write down your insights

```python
a= df.groupby("city").count()["id"]
a
b= a.sort_values(ascending=False).index[0]
b
```

Most matches were played in mumbai(85).

```python
a.plot(kind='bar')
plt.xlabel("City")
plt.ylabel("Number of Matches")
plt.title("Number of Matches hosted by Cities")
plt.show()
```

According to the stats, Most matches were held in mumbai that are 85 matches.Above bar chart also represents the which city hosted most number of matches

# Find all venue of mumbai city

```python
v=df[df["city"] == "Mumbai"]["venue"].unique()
v
```

# now compare in which venue of mumbai most number of matches played (draw bar plot and  write down insights)

```python
c=df[df["city"] == "Mumbai"]
d= c.groupby("venue").count()["id"]
e= d.sort_values(ascending=False).index[0]
e
```

```python
d.plot(kind='bar')
plt.xlabel("Venue")
plt.ylabel("Number of Matches")
plt.title("Number of Matches played in different Venues of Mumbai")
plt.show()
```

By this we come to know that in the venue of mumbai most number of matches played were in Wankhede stadium.

# What is the preferred choice after winning a toss in mumbai

```python
f=df[(df["city"] == "Mumbai") & (df["toss_decision"] != "field")] 
g=f["toss_decision"].value_counts()
g
```

```python
g.plot(kind='pie',autopct='%1.1f%%')
plt.title("Preferred choice after winning a toss in Mumbai")
plt.show()
```

# lets analyse ipl season held in 2013

```python
u = df[(df["season"] == 2013) ]
v= u["city"].value_counts()
plt.pie(v.values, labels = v.index, autopct='%1.1f%%', radius=10, labeldistance=2.6, shadow=True, frame=True, normalize=True)
plt.title("Number of matches hosted by each city in 2013")
plt.legend(prop={'size': 6})
plt.show()
```
